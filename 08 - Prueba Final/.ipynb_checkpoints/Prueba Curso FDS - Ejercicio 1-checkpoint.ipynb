# Modelación descriptiva

Se utilizará la función `saturated_model` para observar cuáles variables aportan mayor explicación a la variable objetivo `income`

```python
modelo_saturado = fnc.saturated_model(df_nonan,'income')
```

```python
modelo_saturado.summary().tables[1]
```

```python
fnc.coefplot(modelo_saturado, varnames=True, intercept=False, fit_stats=True, figsize=(7, 12))
```

En base a las variables que tengan un $p-value < 0.05$ se correra un modelo depurado para comparar los resultados con el modelo saturado.


```python
modelo_saturado.params[modelo_saturado.pvalues < 0.05]
```

```python
modelo_depurado=smf.logit('income ~ age+capital_gain+capital_loss+hours_per_week+sex+not_in_family+own_child+unmarried+wife+other_relative+black+asian_pac_islander+amer_indian_eskimo+blue_collar+others_occup+self_employed+federal_gov+never_married+divorced+separated+widowed+college+university+elementary_school',data=df_nonan).fit()
```

```python
modelo_depurado.summary().tables[1]
```

#### Estadísticos de bondad de ajuste entre ambos modelos

```python
fnc.stats_bondad_ajuste(modelo_saturado)
```

```python
fnc.stats_bondad_ajuste(modelo_depurado)
```

#### Conclusiones modelos descriptivos

El estadístico de `LLR p-value` indica cuan cercano está el modelo para explicar la variable estudiada, por lo cual, el `modelo depurado` explica marginalmente de mejor manera el comportamiento de la variable `income`.

El modelo depurado mejora marginalmente vs el modelo saturado (dif entre Log -Lighlihood vs null no disminuyó sustancialmente) y los coeficientes se mantienen relativamente similares, por lo que el modelo es estable y consistente.

#### Medición de probabilidad con inverse logit

Al utilizar un modelo de regresión logistica, no se pueden interpretar directamente los coeficientes del modelo (reflejado en los logodds), por lo cual, es necesario calcular la probabilidad a través de la función de inversa de logaritmo con `inverse_logit`.

```python
modelo_pred_1 = modelo_saturado
modelo_pred_2 = modelo_depurado
```

```python
m1_estimate_y = 0
for index in modelo_pred_1.params.drop('Intercept').index:
    m1_estimate_y += modelo_pred_1.params[index] * df_nonan[index].mean()
m1_estimate_y += modelo_pred_1.params['Intercept']
```

```python
m2_estimate_y = 0
for index in modelo_pred_2.params.drop('Intercept').index:
    m2_estimate_y += modelo_pred_2.params[index] * df_nonan[index].mean()
m2_estimate_y += modelo_pred_2.params['Intercept']
```

```python
fnc.inverse_logit(m1_estimate_y)
```

```python
fnc.inverse_logit(m2_estimate_y)
```

**Para el modelo1**: Un individuo con características similares a la muestra depurada tiene aproximadamente un 13,43% de probabilidad de tener más de 50K de income

**Para el modelo2**: Un individuo con características similares a la muestra depurada tiene aproximadamente un 13,53% de probabilidad de tener más de 50K de income

1. Modelo con variables seleccionadas a partir del modelo_depurado sin modificación a los hiperparámetros
2. Modelo con variables selecionadas a partir del modelo_saturado sin modificación a los hiperparámetros

#### Métricas de desempeño para las tareas de clasificación con Matriz de Confusión

Una de las manera para medir el desempeño del modelo, es a través de la matriz de confusión, la cual nos permite observar la cantidad de observaciones predichas de forma correcta. 

La matriz de confusión, se puede interpretar de la siguiente manera:

| Predicción/Actuales | Positivo | Negativo |
|----|---|---|
| Positivo | Verdadero Positivo | Falso Positivo |
| Negativo | Falso Negativo | Verdadero Negativo |

Esta labor se puede realizar con el método `confusion_matrix` de `sklearn` en su módulo `metrics` el cual permite desarrollar el cruce de información, o de manera alternativa con `pd.crosstab()`. Este método pide como argumentos el vector objetivo de prueba (`y_test`) y el vector de etiquetas predichas (`yhat`).

```python
m1_conf_pd = pd.crosstab(y_test_m1, yhat_m1)
m2_conf_pd = pd.crosstab(y_test_m2, yhat_m2)
m3_conf_pd = pd.crosstab(y_test_m1, yhat_m3)
m4_conf_pd = pd.crosstab(y_test_m2, yhat_m4)
```

#### Receiving Operating Characteristics (Curva ROC)

* Con _Precision_, _Recall_ y _F1_ logramos caracterizar las tasas de predicción exitosa del modelo. 
* En las matrices de confusión se reportan dos tipos de errores que todo problema de clasificación debe tomar en cuenta:
    - Un __falso positivo__ que surge cuando el modelo estima $\hat{y}=1$ pero la clasificación verdadera es $y=0$
    - Un __falso negativo__ que surge cuando el modelo estima $\hat{y}=0$ pero la clasificación verdadera es $y=1$.
    
* Un buen modelo de clasificación también debe considerar cuál es el rango de errores en el modelo con el que se trabaja. Para ello utilizamos la curva ROC (_Receiving Operating Characteristics_), que evalúa la relación entre ambos errores condicional en todo el rango del clasificador.

* Para generar un gráfico de curva ROC con `sklearn`, utilizamos el método `roc_curve` dentro del módulo `metrics`. El método devuelve tres objetos: Un array con la tasa de falsos positivos, un array con la tasa de verdaderos positivos y un array con los umbrales de decisión entre 0 y 1.

```python
# reestimamos los valores predichos de nuestro modelo para obtener la probabilidad entre 0 y 1.
yhat_m1 = modelo_pred_1.predict_proba(X_test_m1)[:, 1]
yhat_m2 = modelo_pred_2.predict_proba(X_test_m2)[:, 1]
yhat_m3 = modelo_pred_3.predict_proba(X_test_std_m3)[:, 1]
yhat_m4 = modelo_pred_4.predict_proba(X_test_std_m4)[:, 1]

# generamos los objetos de roc_cruve
false_positive_1, true_positive_1, threshold_1 = roc_curve(y_test_m1, yhat_m1)
false_positive_2, true_positive_2, threshold_2 = roc_curve(y_test_m2, yhat_m2)
false_positive_3, true_positive_3, threshold_3 = roc_curve(y_test_m1, yhat_m3)
false_positive_4, true_positive_4, threshold_4 = roc_curve(y_test_m2, yhat_m4)
```

El gráfico tienen las siguientes convenciones:
* En el eje Y va tasa de __falsos positivos__, aquella que resume las falsas alarmas en nuestro modelo clasificador.
* En el eje X va la tasa de __verdaderos positivos__, aquella que resume las observaciones correctamente clasificadas por nuestro modelo.
* La línea bisectriz representa el __benchmark__ predictivo del modelo: esto es lo que esperamos si es que un clasificador asignara etiquetas de forma aleatoria, sin mayor información ingresada por el sistema.

```python
# Plot ROC curve
plt.figure(figsize=(15,8))
plt.title('Curva ROC')
plt.plot(false_positive_1, true_positive_1, lw=1, label='modelo_1')
plt.plot(false_positive_2, true_positive_2, lw=1, label='modelo_2')
plt.plot(false_positive_3, true_positive_3, lw=1, label='modelo_3')
plt.plot(false_positive_4, true_positive_4, lw=1, label='modelo_4')
plt.plot([0, 1], ls="--", lw=1)
plt.plot([0, 0], [1, 0] , c='limegreen', lw=3), plt.plot([1, 1] , c='limegreen', lw=3)
plt.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1, facecolor='white', fontsize='medium')
plt.ylabel('Verdaderos Positivos')
plt.xlabel('Falsos Positivos');
```

Los modelos presentan una curva por sobre la bisectriz, lo que indica que tienen una capacidad predictiva adecuada, esto se interpreta que la tasa de verdaderos positivos es mayor que la tasa de falsos positivos.

La calidad de la curva ROC también se puede resumir con una métrica llamada la __área debajo de la curva__. Los Valores mayores de ésta cifra van a estar asociados con un mejor desempeño del modelo en predecir más verdaderos positivos que falsos positivos.

 Para estimar el área debajo de la curva utilizamos el método `roc_auc_score` en el módulo `metrics` de `sklearn`. De manera similar a las implementaciones de métricas previas con `sklearn`, el método solicita el vector objetivo verdadero (`y_test`) y nuestras predicciones en base al modelo (`yhat`).

```python
print("modelo_1 : ", roc_auc_score(y_test_m1, yhat_m1))
print("modelo_2 : ", roc_auc_score(y_test_m2, yhat_m2))
print("modelo_3 : ", roc_auc_score(y_test_m1, yhat_m3))
print("modelo_4 : ", roc_auc_score(y_test_m2, yhat_m4))
```

El resultado del AUC se puede interpretar que para el mejor modelo (__modelo_4__) existe una probabilidad de __89,99%__ de que una instancia positiva elegida al azar tenga una clasificación más alta que una instancia negativa elegida al azar

#### k-Fold Cross Validation (Validación Cruzada repetida _k_ veces)

* La técnica de Cross Validadtion nos permite separar de forma repetida en múltiples muestras de entrenamiento y validación los datos del modelo.
* Por cada muestra se calcula una métrica de desempeño y posteriormente son promediadas para tener una aproximación al costo asociado de implementar un modelo.
* Para implementar una validación cruzada con `sklearn` importamos el método `cross_val_score` en el módulo `model_selection`. A diferencia de la segmentación de muestras clásica realizada con el método `train_test_split` donde se define el porcentaje a asignar en la muestra de validación, con este método se define la cantidad de submuestras a considerar.
* Dado que el modelo_4 fue el que presentó mejor métrica de AUC, se utilizará para generar la validación cruzada.

```python
m4_cv = cross_val_score(LogisticRegression(), 
                        X=StandardScaler().fit_transform(df_nonan.drop(columns='income', axis=1)),
                        y=df_nonan['income'],
                        cv=10, 
                        scoring='f1')
```

* El método `cross_val_score` devuelve un array con la métrica parametrizada para cada submuestra, en este caso se solicitó el `F1`  .

```python
for i, n in enumerate(m4_cv):
    print("El puntaje F1 para la muestra {0} es {1}".format(i+1, round(n, 2)))
```

```python
print("El puntaje F1 promedio para el modelo es de: ", round(np.mean(m4_cv), 2))
print("La desviación estandar F1 promedio para el modelo es de: ", round(np.std(m4_cv),2))
```

* Cuando promediamos los puntajes, la evaluación del modelo se mantiene relativamente estable alrededor de un 65% con una desviación estandar baja. En comparación a nuestro F1 original del modelo sin validación cruzada, observamos que no existe cambio para esta métrica.

```python
# definimos los criterios en un array
k_fold = [3, 5, 10, 20]
# generamos un gráfico vacío que contenga los gráficos 
fig, ax = plt.subplots(1, len(k_fold), sharey='row')
# por cada elemento y su posición en el array
for n, i in enumerate(k_fold):
    # estimar la validación cruzada con el elemento
    tmp = cross_val_score(LogisticRegression(), 
                        X=StandardScaler().fit_transform(df_nonan.drop(columns='income', axis=1)),
                        y=df_nonan['income'], 
                        cv = i,
                        scoring='f1')
    # guardar el gráfico en la posición declarada
    ax[n].boxplot(tmp)
    ax[n].set_title('k={0}'.format(i))

fig.suptitle('Desempeño del modelo logístico (Puntaje F1) condicional a la cantidad de validaciones cruzadas');
```

* En el gráfico creado se visualiza que el riesgo de utilizar un número alto de segmentaciones (k=20) implica aumentar la variablidad de nuestra métrica de desempeño y levanta dudas sobre la utilidad del modelo para desarrollar nuevas predicciones.
* Cuando el número de validaciones es bajo (k=3), el riesgo es en considerar que la métrica promediada es la óptima. Disminuímos la variabilidad de la predicción a expensas de aumentar el sesgo y obtener un parámetro que dista del óptimo empírico.

#### Refactorización del modelo

A partir del `modelo_4`, se generarán tres refactorizaciones del modelo:
1. __Modelo lineal aditivo:__ Este es el modelo que desarrollamos anteriormente. Se dice que es _aditivo_ dado que los 4 atributos ingresados no interactúan entre sí.
2. __Modelo lineal con interacciones de 2-orden:__ De forma adicional a los parámetros asignados en el modelo lineal aditivo, agregamos 6 términos que representan interacciones entre dos variables.

#### Interacciones del modelo

* Para generar las interacciones utilizaremos la clase `PolynomialFeatures()` dentro del módulo `preprocessing` de `sklearn`. Dentro de los parámetros de la clase declaramos las siguientes especificaciones:

    * `degree=2`: Restringimos que el orden de las interacciones no debe superar 2.
    * `interaction_only=True`: Con esta opción ignoramos los términos cuadráticos que se generan por defecto, en adición a las interacciones.
    * `include_bias=False`: Para conservar grados de libertad, ignoramos la estimación del intercepto.
    
* Posteriormente, concatenamos la función `fit_transform` a la clase. Dentro de ella pasamos __las matrices de atributos estandarizadas para entrenamiento y validación__.

```python
# estandarizamos la matriz de atributos
std_matrix = StandardScaler().fit_transform(df_nonan.drop(columns='income', axis=1))

# Con la clase PolynomialFeature preparamos las otras dos matrices candidatas
from sklearn.preprocessing import PolynomialFeatures

# interacción de 2do orden
df_int_2 = PolynomialFeatures(degree=2, 
                              interaction_only=True, 
                              include_bias=False).fit_transform(std_matrix)
```

#### Comparaciones de múltiples métricas

Se implementará un flujo para graficar el comportamiento de las métricas _F1_, _Precission_, _Recall_ y _Accuracy_ en los tres modelos mediante validación cruzada, estas métricas serán guardadas en un array (`eval_metrics`) el cual determinará el loop a realizar.

Se generará 4 gráficos de caja, uno por cada métrica, solicitando lo siguiente:
   1. Calcula los puntajes de validación cruzada con `k=3` para cada modelo candidato.
   2. Genera un objeto `DataFrame` para guardar las métricas de forma _larga_ mediante `.unstack()`.
   3. Por cada objeto `DataFrame`, graficamos el diagrama de caja.

4. Finiquitamos con detalles estéticos de presentación.

```python
eval_metrics = ['f1', 'precision', 'recall', 'accuracy']

for n, i in enumerate(eval_metrics):
    tmp_1 = cross_val_score(LogisticRegression(), 
                        X=StandardScaler().fit_transform(df_nonan.drop(columns='income', axis=1)),
                        y=df_nonan['income'],
                        cv=3, 
                        scoring=i)
    tmp_2 = cross_val_score(LogisticRegression(), 
                        X=df_int_2,
                        y=df_nonan['income'],
                        cv=3, 
                        scoring=i)
    
    tmp = pd.DataFrame({'Additive': tmp_1,
                        '2-order Interaction': tmp_2}).unstack().reset_index()
    tmp.rename(columns={'level_0':'order',
                        'level_1': 'num',
                        '0': 'score'}, inplace=True)
    plt.subplot(1, 4, n+1)
    sns.boxplot(tmp.iloc[:,0],
                tmp.iloc[:, 2])
    sns.despine()
    plt.xticks(rotation=45, ha='right')
    plt.xlabel(' ')
    plt.ylabel(' ')
    plt.title(i.capitalize())
    plt.tight_layout;
```

#### Selección del mejor modelo

Si analizamos los resultados obtenidos de las distintas métricas, podemos observamos que el __Modelo lineal aditivo__ tiene mejor desempeño que el iteraciones del segundo orden sin embargo, presenta una mayor varianza que el modelo con interacciones del segundo orden.

#### Experimento: ¿Mayor complejidad del modelo conlleva Mejores resultados?

```python
# estandarizamos la matriz de datos
std_matrix = StandardScaler().fit_transform(df_nonan.drop(columns='income', axis=1))
#declaramos los términos
polynomial_order = [1, 2, 3]
# por cada elemento del array
for i in polynomial_order:
    
    #generamos muestras segmentadas
    X_train_tmp, X_test_tmp, y_train_tmp, y_test_tmp = train_test_split(std_matrix,
                                               df_nonan['income'],
                                               test_size=.33,
                                               random_state=250992)
    
    # Transformamos la muestra de entrenamiento por i polinomio
    X_train_tmp = PolynomialFeatures(degree=i,
                                     interaction_only=False,
                                     include_bias=False).fit_transform(X_train_tmp)
    # Transformamos la muestra de evaluación por i polinomio
    X_test_tmp = PolynomialFeatures(degree=i,
                                    interaction_only=False,
                                    include_bias=False).fit_transform(X_test_tmp)
    # Generamos el modelo predictivo
    tmp_model = LogisticRegression().fit(X_train_tmp, y_train_tmp)
    # Guardamos las predicciones de probabilidad
    tmp_pred = tmp_model.predict_proba(X_test_tmp)[:,1]
    # generamos los arrays con roc_curve
    false_positive, true_positive, threshold = roc_curve(y_test_tmp,
                                                         tmp_pred)
    # redondeamos los valores del puntaje AUC
    roc_auc_tmp = round(roc_auc_score(y_test_tmp, tmp_pred), 3)
    # graficamos
    plt.plot(false_positive,
             true_positive,
             label = 'Orden: {0} - AUC: {1}'.format(i, roc_auc_tmp))
    plt.plot([0, 1], ls="--")
    plt.plot([0, 0], [1, 0] , c='limegreen', lw=3),
    plt.plot([1, 1] , c='limegreen', lw=3)
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.legend()
    # solicitamos información de iteración
    print("Orden: {0}, cantidad de atributos estimables: {1}. AUC estimada: {2}".format(i, X_train_tmp.shape[1], roc_auc_tmp)) 
```

#### Conclusiones
* El output del entrenamiento de los modelos demuestra que el aumentar la cantidad de parámetros, no necesariamente conlleva a un aumento del desempeño del modelo, ya que se alcanzó un máximo del AUC de 0.905 para el modelo con poliminomios de 2-orden. 
* Adicionalmente, es necesario cuestionarse si el _trade_ de rendimiento/resultados es justificado, ya que el tiempo de procesamiento para un modelo polinomial con orden $>=3$ es considerablemente alto en comparación con el entramiento de un modelo de regresión logistica con hiperparámetros modificados.
* Para este experimiento, si comparamos la métrica del AUC del modelo saturado inicialmente entrenado con el modelo polinomial de 2-orden, podemos observar que obtuvimos un _upside_ de 0.006 pts en la medida de AUC para el modelo polinomial de 2-orden. 
* Finalmente cabe señalar que el modelo sin términos polinomiales tiene un desempeño relativamente adecuado (comparado entre todos los modelos candidatos).
