## Transformación, recodificación y limpieza de datos

Se binarizarán las variables del dataset, siguiendo el criterio de asignarle 1 a las categorías minoritarias

```python
df_nonan['school'] = np.where(df_nonan['school'] == 'MS', 1, 0)
df_nonan['sex'] = np.where(df_nonan['sex'] == 'M', 1, 0)
df_nonan['address'] = np.where(df_nonan['address'] == 'R', 1, 0)
df_nonan['famsize'] = np.where(df_nonan['famsize'] == 'LE3', 1, 0)
df_nonan['Pstatus'] = np.where(df_nonan['Pstatus'] == 'A', 1, 0)
df_nonan['schoolsup'] = np.where(df_nonan['schoolsup'] == 'yes', 1, 0)
df_nonan['famsup'] = np.where(df_nonan['famsup'] == 'no', 1, 0)
df_nonan['paid'] = np.where(df_nonan['paid'] == 'yes', 1, 0)
df_nonan['activities'] = np.where(df_nonan['activities'] == 'no', 1, 0)
df_nonan['nursery'] = np.where(df_nonan['nursery'] == 'no', 1, 0)
df_nonan['higher'] = np.where(df_nonan['higher'] == 'no', 1, 0)
df_nonan['internet'] = np.where(df_nonan['internet'] == 'no', 1, 0)
df_nonan['romantic'] = np.where(df_nonan['romantic'] == 'yes', 1, 0)
```

```python
df_nonan['Mjob']=df_nonan['Mjob'].replace('other', 'mjob_other')
df_nonan['Mjob']=df_nonan['Mjob'].replace('services', 'mjob_services')
df_nonan['Mjob']=df_nonan['Mjob'].replace('teacher', 'mjob_teacher')
df_nonan['Mjob']=df_nonan['Mjob'].replace('at_home', 'mjob_at_home')
df_nonan['Mjob']=df_nonan['Mjob'].replace('health', 'mjob_health')

df_nonan['Fjob']=df_nonan['Fjob'].replace('other', 'fjob_other')
df_nonan['Fjob']=df_nonan['Fjob'].replace('services', 'fjob_services')
df_nonan['Fjob']=df_nonan['Fjob'].replace('teacher', 'fjob_teacher')
df_nonan['Fjob']=df_nonan['Fjob'].replace('at_home', 'fjob_at_home')
df_nonan['Fjob']=df_nonan['Fjob'].replace('health', 'fjob_health')

df_nonan['reason']=df_nonan['reason'].replace('course', 'reason_course')
df_nonan['reason']=df_nonan['reason'].replace('reputation', 'reason_reputation')
df_nonan['reason']=df_nonan['reason'].replace('home', 'reason_home')
df_nonan['reason']=df_nonan['reason'].replace('other', 'reason_other')

df_nonan['guardian']=df_nonan['guardian'].replace('other', 'guardian_other')
df_nonan['guardian']=df_nonan['guardian'].replace('mother', 'guardian_mother')
df_nonan['guardian']=df_nonan['guardian'].replace('father', 'guardian_father')
```

# Dimensionalidad y Agrupación
#### Prueba de esferacidad de errores de Bartlett

Es necesario estimar si la matriz de correlación del dataset es  una matriz de identidad, para realizar esta labor utilizaremos el método `calculate_bartlett_sphericity` de la librería `factor_analyzer`

```python
fact.calculate_bartlett_sphericity(df_nonan)
```

La prueba retorna un puntaje `chi-square` de la distribución y un `p-value` asociado a esa prueba. En este caso, dado que el p-value bastante menor a 0.05, se sugiere que la matriz de correlaciones no es igual a una matriz de identidad, por lo que se puede seguir con el análisis de factores.

#### Prueba de muestreo Kaiser-Meyer-Olkin (KMO)

* De manera adicional a la prueba de Bartlett, podemos medir si las correlaciones parciales dentro de la matriz son cercanas a 0. De ser este el caso, la prueba sugiere la existencia de por lo menos un factor latente dentro del dataset.
* Se puede implementar mediante el método `calculate_kmo` de la librería `factor_analyzer`.

```python
fact.calculate_kmo(df_nonan)
```

El método indica la proporción de varianza en sus variables que puede ser causada por factores subyacentes a través del puntaje KMO para cada item dentro del dataset, así como un puntaje global para todos los items.

En este caso, la prueba de muestreo KMO indica que existen atributos que tienen valores bajo 0.7, por lo cual, __no se debe continuar con el Análisis Factorial.__

### Modelo descriptivo de G3

#### Refactorización del modelo con Feature engineering

* Los modelos generados con `sklearn` incluye todas las variables dentro de la matriz de datos. Un inconveniente relacionado es el hecho que nuestro modelo puede que no sea el óptimo, dado que incluimos regresores que no aportan poder predictivo y utilizan grados de libertad de forma ineficiente. 

* El objetivo de la predicción es encontrar un número adecuado de regresores a incluír en nuestro modelo predictivo.

* Una manera simple de realizar este labor es mediante la generación de correlaciones parciales entre cada atributo y el vector objetivo. 

```python
# extraemos los nombres de las columnas en la base de datos
columns = df_nonan.columns

# generamos 3 arrays vacíos para guardar los valores
# nombre de la variable
attr_name = []
# correlación de pearson
pearson_r = []
# valor absoluto de la correlación
abs_pearson_r = []

# para cada columna en el array de columnas
for col in columns:
    # si la columna no es la dependiente
    if col != "G3":
        # adjuntar el nombre de la variable en attr_name
        attr_name.append(col)
        # adjuntar la correlación de pearson
        pearson_r.append(df_nonan[col].corr(df_nonan['G3']))
        # adjuntar el absoluto de la correlación de pearson
        abs_pearson_r.append(abs(df_nonan[col].corr(df_nonan['G3'])))
        
# transformamos los arrays en un DataFrame
features = pd.DataFrame({
    'attribute': attr_name,
    'corr':pearson_r,
    'abs_corr':abs_pearson_r
})

# generamos el index con los nombres de las variables
features = features.set_index('attribute')
# ordenamos los valores de forma descendiente
features.sort_values(by=['abs_corr'], ascending=False)
```

#### Implementando un modelo con menos regresores

* En base a la información obtenida con nuestra tabla de correlaciones parciales, podemos refinar nuestro primer modelo para reducir el error cuadrático.
* Para ello seleccionaremos aquellas correlaciones que sean superior al .30. 

```python
features[features['abs_corr']>0.3]
```

Las variables que cumplen con esta condición son `failures`, `G1` y `G2`

```python
# separemos los vectores a trabajar
y_vec= df_nonan.loc[:, 'G3']
X_mat = df_nonan.loc[:, ['failures', 'G1', 'G2']]
```

* Dividiremos la muestra en conjuntos de entrenamiento y prueba utilizando el método `train_test_split`,el cual permite realizar separaciones de muestras aleatorizadas. 

* Cabe destacar que éste método es el preferido, dado que mediante la aleatorización podemos cancelar el efecto de sesgo por variables intervinientes en la muestra.

* El método toma como argumentos la matriz de atributos (`X_mat`) y el vector objetivo a dividir (`y_vec`), el porcentaje de datos que serán asignados al conjunto de muestra (`test_size`), y una semilla pseudoaleatoria para asegurar replicación de resultados (`random_state`).

* `train_test_split` devuelve como objetos la matriz de atributos para entrenar (`X_train`), la matriz de atributos para validar (`X_test`), el vector objetivo para entrenar (`y_train`) y el vector objetivo para contrastar las predicciones (`y_test`). 

```python
X_train, X_test, y_train, y_test = train_test_split(X_mat, y_vec, test_size=.30, random_state=250992)
```

* Con la muestra ya dividida, podemos implementar el modelo de regresión con los mismos hiperparámetros anteriores.

```python
purge_model = linear_model.LinearRegression(fit_intercept=True, normalize=True)
purge_model.fit(X_train, y_train)
purge_model_yhat = purge_model.predict(X_test)
```

* Ahora podemos comparar el poder predictivo entre los dos modelos mediante el promedio del error cuadrático y el r cuadrado. 

```python
m1_mse = mse[3]
m2_mse = mean_squared_error(y_test, purge_model_yhat).round(0)

m1_r2 = r2[3]
m2_r2 = r2_score(y_test, purge_model_yhat).round(2)

plt.figure(figsize=(10,5))
plt.subplot(2,1,1)
plt.barh(['Modelo 1\nTodos los Atributos', 'Modelo 2\nEd + Male + Height'], [m1_mse, m2_mse])
plt.title("Promedio Error Cuadrático")
plt.subplot(2,1, 2)
plt.barh(['Modelo 1\nTodos los Atributos', 'Modelo 2\nEd + Male + Height'], [m1_r2, m2_r2])
plt.title("R cuadrado");
plt.tight_layout()
```

__Conclusiones__:

* Si comparamos el promedio del error cuadrático para ambos modelos, existe evidencia para preferir el modelo con menos regresores, dado que presenta una reducción significativa del error cuadrático.

* De igual manera sucede con el estadístico $R^{2}$, que presenta una mayor cercanía de los datos a la línea de regresión ajustada
