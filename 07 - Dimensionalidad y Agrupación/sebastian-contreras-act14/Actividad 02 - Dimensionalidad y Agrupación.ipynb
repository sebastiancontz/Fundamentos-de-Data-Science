```python
alto, ancho, canal = tmp.shape
```

```python
datos_procesados = tmp.reshape(alto * ancho, canal)
```

```python
datos_procesados.shape
```

```python
%%time
kmeans = KMeans(n_clusters=16).fit(datos_procesados)
```

```python
kmeans.cluster_centers_
```

```python
new_colors_pred = kmeans.cluster_centers_[kmeans.predict(datos_procesados)].astype(int)
```

```python
def rgb_hex_it(rgb):
    tmp = list(map(lambda x: int(x), rgb))
    return "#%02x%02x%02x" % (tmp[0], tmp[1], tmp[2])
```

```python
hex_convert = list(map(lambda x: rgb_hex_it(x), new_colors_pred))
unique, count = np.unique(kmeans.labels_, return_counts=True)
```
