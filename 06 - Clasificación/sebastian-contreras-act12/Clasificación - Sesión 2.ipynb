* A partir del modelo depurado, genere un modelo predictivo con las siguientes características:

- Implemente el modelo sin alterar los hiperparámetros.

* En base a los pasos, comente sobre el desempeño predictivo del modelo.

```python
X_train, X_test, y_train, y_test = train_test_split(df.loc[:, ['balance', 'student']],
                                                   df['default'],
                                                   test_size=.33,
                                                   random_state=250992)
```

```python
default_model = LogisticRegression().fit(X_train, y_train)
```

```python
yhat = default_model.predict(X_test)
```

```python
m1_conf_pd = pd.crosstab(y_test, yhat)
m1_conf_pd
```

Interpretación matriz de confusión:
* Verdadero Positivo  (TP) = 3185
* Falso Positivo (FN) = 11
* Falso Negativo (FP) = 77
* Verdadero Negativo (TN) = 27


| Predicción/Actuales | Positivo | Negativo |
|----|---|---|
| Positivo | 3185 | 11 |
| Negativo | 77 | 27 |

```python
m1_prec = precision_score(y_test, yhat)
m1_rec = recall_score(y_test, yhat)
print("Precision: ", m1_prec, "\nRecall: ", m1_rec)
```

* El indicador de `Precision` indica que un 71,05% de identificaciones positivas fue correcta
* El indicador de `Recall` indica que un 25,96% de positivos reales se identificó correctamente 

```python
print(classification_report(y_test, yhat))
```

A partir de la tabla `F1` se pueden generar las siguientes observaciones:
* Para $y_{i}=0$ el modelo presenta métricas por sobre el 90%
* Para $y_{i}=1$ el modelo presenta métricas cercanas al 38%

En base a estos resultados, se puede apreciar que el modelo tiene problemas para clasificar correctectamente los casos donde $y_{i}=1$.
Esto sugiere que el modelo debe ser refactorizado y encontrar predictores que estén asociados con los casos donde $y_{i}=1$.

```python
# reestimamos los valores predichos de nuestro modelo para obtener la probabilidad entre 0 y 1.
yhat = default_model.predict_proba(X_test)[:, 1]

# generamos los objetos de roc_cruve
false_positive, true_positive, threshold = roc_curve(y_test, yhat)
```

```python
# Plot ROC curve
plt.title('Curva ROC')
plt.plot(false_positive, true_positive, lw=1)
plt.plot([0, 1], ls="--", lw=1)
plt.plot([0, 0], [1, 0] , c='limegreen', lw=3), plt.plot([1, 1] , c='limegreen', lw=3)
plt.ylabel('Verdaderos Positivos')
plt.xlabel('Falsos Positivos');
```

```python
roc_auc_score(y_test, yhat)
```

El resultado del AUC se puede interpretar que existe una probabilidad de 94,93% de que una instancia positiva elegida al azar tenga una clasificación más alta que una instancia negativa elegida al azar.
